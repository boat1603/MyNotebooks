### Merging, Joining, and Concatenating 

##### Merging

```python
df_left=pd.DataFrame({'key':'K0 K1 K2 K3'.split(),'A':'A0 A1 A2 A3'.split(),'B':'B0 B1 B2 B3'.split()})
df_right=pd.DataFrame({'key':'K0 K1 K2 K3'.split(),'C':'C0 C1 C2 C3'.split(),'D':'D0 D1 D2 D3'.split()})
```

```python
# Using key
pd.merge(df_left, df_right, how='outer',on='key')
#how={‚Äòleft‚Äô, ‚Äòright‚Äô, ‚Äòouter‚Äô, ‚Äòinner‚Äô}, default ‚Äòinner‚Äô
#left:‡∏£‡∏ß‡∏°‡πÇ‡∏î‡∏¢‡∏¢‡∏∂‡∏îdf ‡∏ó‡∏≤‡∏á‡∏ã‡πâ‡∏≤‡∏¢‡πÄ‡∏õ‡πá‡∏ô‡∏´‡∏•‡∏±‡∏Å  (SQL left outer join)
#right:‡∏£‡∏ß‡∏°‡πÇ‡∏î‡∏¢‡∏¢‡∏∂‡∏îdf‡∏ó‡∏≤‡∏á‡∏Ç‡∏ß‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏´‡∏•‡∏±‡∏Å(SQL right outer join
#outer:‡∏£‡∏ß‡∏°‡πÅ‡∏ö‡∏ö union ‡πÇ‡∏î‡∏¢‡πÄ‡∏≠‡∏≤key ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏°‡∏≤(SQL full outer join)
#inner:‡∏à‡∏∞‡∏£‡∏ß‡∏°‡πÅ‡∏ö‡∏ö‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏â‡∏û‡∏≤‡∏∞ key ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á df
```

```python
df_left=pd.DataFrame({'A':'A0 A1 A2 A3'.split(),'B':'B0 B1 B2 B3'.split(),'key1':'K0 K0 K1 K2'.split(),
                     'key2':'K0 K1 K0 K1'.split()})
df_right=pd.DataFrame({'C':'C0 C1 C2 C3'.split(),'D':'D0 D1 D2 D3'.split(),'key1':'K0 K1 K1 K2'.split(),
                     'key2':'K0 K0 K0 K0'.split()})
print(df_left)
print(df_right)
```

```python
pd.merge(df_left, df_right, how='inner',on=['key1','key2'])
```

```python
pd.merge(df_left, df_right, how='outer',on=['key1','key2'])
```

```python
pd.merge(df_left, df_right, how='left',on=['key1','key2'])
```

```python
pd.merge(df_left, df_right, how='right',on=['key1','key2'])
```

##### Joining 

```python
df_left=pd.DataFrame({'A':'A0 A1 A2'.split(),'B':'B0 B1 B2'.split()},index='K0 K1 K2'.split())
df_right=pd.DataFrame({'C':'C0 C2 C3'.split(),'D':'D0 D2 D3'.split()},index='K0 K2 K3'.split())
```

```python
#Using index to join
df_left.join(df_right)
```

```python
df_left.join(df_right,how='outer')
```

##### Concatenating

```python
df01=pd.DataFrame({'A':'A0 A1 A2 A3'.split(),'B':'B0 B1 B2 B3'.split(),'C':'C0 C1 C2 C3'.split(),
                 'D':'D0 D1 D2 D3'.split()},index=[0,1,2,3])
df02=pd.DataFrame({'A':'A4 A5 A6 A7'.split(),'B':'B4 B5 B6 B7'.split(),'C':'C4 C5 C6 C7'.split(),
                 'D':'D4 D5 D6 D7'.split()},index=[4,5,6,7])
df03=pd.DataFrame({'A':'A8 A9 A10 A11'.split(),'B':'B8 B9 B10 B11'.split(),'C':'C8 C9 C10 C11'.split(),
                 'D':'D8 D9 D10 D11'.split()},index=[8,9,10,11])
```

```python
pd.concat([df01,df02,df03])
```

```python
pd.concat([df01,df02,df03],axis=1)
```
